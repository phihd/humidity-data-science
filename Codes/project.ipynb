```python
import pandas as pd
import re
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
```

```python
X_train = pd.read_csv('weather_data_train.csv', index_col = 'datetime', sep = ';', decimal = ',', infer_datetime_format = True)
X_train.head(10)
```

```python
#print(X_train.Tn_mu)
print(min(X_train.Tn_mu))
```

```python
plt.figure()
plt.hist(X_train.Tn_mu, bins = 'auto', edgecolor='k')
plt.title('Histogram plot of the the mean of the minimum air temp')
plt.xlabel('Celsius')
plt.ylabel('Frequency')
plt.xticks(range(int(min(X_train.Tn_mu)) - 1, int(max(X_train.Tn_mu)) + 1,5))
plt.show()
```

```python
plt.figure()
plt.hist(X_train.Tx_mu, bins = 'auto', edgecolor='k')
plt.title('Histogram plot of the the mean of the maximum air temp')
plt.xlabel('Celsius')
plt.ylabel('Frequency')
plt.xticks(range(int(min(X_train.Tx_mu)) - 1, int(max(X_train.Tx_mu)) + 1,5))
plt.show()
```
